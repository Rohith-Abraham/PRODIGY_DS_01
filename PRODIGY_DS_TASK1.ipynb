```python
inc_group = meta_data['IncomeGroup'].value_counts()
```

```python
plt.figure(figsize=(8, 8))
plt.pie(inc_group, labels=inc_group.index,autopct='%1.1f%%')
plt.title('Distribution of Countries Across Income Groups')
plt.axis('equal')
plt.show()
```

```python
import seaborn as sns
```

```python
region_count = meta_data['Region'].value_counts()
```

```python
plt.figure(figsize=(10, 6))
sns.barplot(x=region_count.index, y=region_count.values, palette="Set3")
plt.xlabel('Region')
plt.ylabel('Number of Countries')
plt.title('Number of Countries in Each Region')
plt.xticks(rotation=30)
plt.show()
```

```python

plt.figure(figsize=(10, 6))
sns.barplot(x=region_count.index, y=region_count.values, hue=region_count.index, palette="Set3", legend=False)
plt.xlabel('Region')
plt.ylabel('Number of Countries')
plt.title('Number of Countries in Each Region')
plt.xticks(rotation=30)
plt.show()
```

```python
region_income = pd.crosstab(meta_data['Region'], meta_data['IncomeGroup'])
```

```python
plt.figure(figsize=(10, 6))
region_income.plot(kind='bar', stacked=True, colormap='plasma')
plt.xlabel('Region')
plt.ylabel('Count')
plt.title('Income Groups Within Each Region')
plt.xticks(rotation=45)
plt.show()
```
